univariate analysis

```python
# summary statistics for sales_cleaned data
numeric_cols = ['quantity', 'total_amount']

summary = data[numeric_cols].describe().T
summary['median'] = data[numeric_cols].median()
summary['IQR'] = summary['75%'] - summary['25%']
summary['skewness'] = data[numeric_cols].skew()

print("Numeric Summary Statistics:\n")
summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# histogram, boxplot, density plot for numerical columns
for col in numeric_cols:
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram of {col}')

    plt.subplot(1, 3, 2)
    sns.boxplot(y=data[col])
    plt.title(f'Boxplot of {col}')

    plt.subplot(1, 3, 3)
    sns.kdeplot(data[col], fill=True)
    plt.title(f'Density Plot of {col}')

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

top10 = merge.groupby("customer_id")["total_amount"] \
             .sum() \
             .sort_values(ascending=False) \
             .head(10)

plt.figure(figsize=(10, 6))

bars = plt.bar(top10.index.astype(str), top10.values, color="#2E86C1")




plt.title("Top 10 Customer IDs by Sales", fontsize=14, fontweight='bold')
plt.xlabel("Customer ID")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)


plt.show()
```
